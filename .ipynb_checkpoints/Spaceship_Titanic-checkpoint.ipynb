# Feature Engineering

Replace the values in the data

```python
def cleanData(data):
    
    data['HomePlanet'].fillna('Earth', inplace=True)
    data['CryoSleep'].fillna(False, inplace=True)    
    data['Destination'].fillna('TRAPPIST-1e', inplace=True)  
    data['VIP'].fillna(False, inplace=True)       
    data['Age'].fillna(24.0, inplace=True)    
    data['RoomService'] = data['RoomService'].transform(lambda x: x.fillna(x.mean()))
    data['FoodCourt'] = data['FoodCourt'].transform(lambda x: x.fillna(x.mean()))    
    data['ShoppingMall'] = data['ShoppingMall'].transform(lambda x: x.fillna(x.mean()))  
    data['Spa'] = data['Spa'].transform(lambda x: x.fillna(x.mean()))    
    data['VRDeck'] = data['VRDeck'].transform(lambda x: x.fillna(x.mean()))     
    
    return data
```

```python
clean_train = cleanData(df_train)
```

```python
clean_train.isna().sum()
```

Let's check whether the data depends on the surname

```python
clean_train['surnames'] = clean_train['Name'].str.split(' ', expand=True)[1]
```

```python
clean_train.query('surnames == "Yorkland"')
```

When checking, I noticed that there are some dependencies of the surname on Transported, but not always. I will not use this data for training

Check cable feature
Takes the form deck/num/side, where side can be either P for Port or S for Starboard.
Let's divide this data

```python
def cleanData2(data):
    
    data.drop(['Name', 'surnames'], axis=1, inplace=True)
    data[['CabinDeck', 'CabinNum', 'CabinSide']] = data['Cabin'].str.split('/', expand=True)
    data.drop(['Cabin'], axis=1, inplace=True)
    
    data['CabinDeck'] = data['CabinDeck'].fillna('F')
    data['CabinSide'] = data['CabinSide'].fillna('S')
    data['CabinNum'] = data['CabinNum'].fillna('82')
    
    return data
```

```python
clean_train = cleanData2(clean_train)
```

```python
# Let's check the values for to fill in the gaps
clean_train.CabinDeck.value_counts(), clean_train.CabinNum.value_counts(), clean_train.CabinSide.value_counts()
```

```python
clean_train.isna().sum()
```

convert some data to columns with 1 or 0

```python
def cleanData3(data):
    
    data = pd.get_dummies(data=data, columns=['HomePlanet', 'Destination', 'CabinSide'], drop_first=True)
    data['CabinDeck'] = label.fit_transform(data['CabinDeck'])
    data['CryoSleep'] = data['CryoSleep'].map({True:1, False:0})
    data['VIP'] = data['VIP'].map({True:1, False:0})
        
    return data
```

```python
clean_train = cleanData3(clean_train)
clean_train['Transported'] = clean_train['Transported'].map({True:1, False:0})
```
